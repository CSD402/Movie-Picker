```python
# Import libraries 
import pandas as pd
import ast
import numpy as np 
```

```python
# Read the data 
movie_data = pd.read_csv('movies_metadata.csv')
```

```python
# Get 5 movies from the data
movie_sample_df = movie_data.sample(5)
```

```python
for i in range(5):    
    movie_genre_list = ast.literal_eval(movie_sample_df.iloc[i]['genres'])[0]
    genre = movie_genre_list['name']
    print(genre)
```
